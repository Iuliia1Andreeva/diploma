```python
import pandas as pd
import numpy as np
df = pd.read_csv('elem_one.csv')
df['ReferenceID'] = -1
df['X1'] = 1
df['El2'] = np.nan
df['X2'] = 0
df['El3'] = np.nan
df['El4'] = np.nan
df['X3'] = 0
df['X4'] = 0
df.rename(columns={'_DateGap': '_date'}, inplace=True)
df.head()
```

```python
df.head()
```

```python
from sqlalchemy import create_engine

# Создаем соединение с базой данных
engine = create_engine("mssql+pyodbc://SA:ANELE$1967_@localhost/YourDatabaseName?driver=ODBC+Driver+17+for+SQL+Server")

# Предположим, что df - ваш DataFrame, а cols - список колонок, которые вы хотите добавить

# Проходим по каждой строке DataFrame
for index, row in df.iterrows():
    try:
        # Выбираем данные из определенных колонок
        selected_data = row[['SubstanceID', 'CrystalSystem', 'SpaceGroup', 'Temperature1', 'Temperature2', 'E1', 'E2', 'IsCalculated', 'ReferenceID']]
        new_df = pd.DataFrame(selected_data).transpose()
        # Добавляем данные в таблицу базы данных
        new_df.to_sql('Gap', con=engine, if_exists='append', index=False)
        print("Данные добавлены успешно.")
    except Exception as e:
        print(f"Ошибка при добавлении данных из строки {index}: {e}")

# Закрываем соединение с базой данных
engine.dispose()
```
