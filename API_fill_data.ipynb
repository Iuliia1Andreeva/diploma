```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
```

```python
elem_one = pd.read_csv('elem_one.csv')
elem_one['ReferenceID'] = -1
elem_one_sql = pd.read_csv('sql_one_elem.csv')
elem_one = pd.concat([elem_one, elem_one_sql], ignore_index=True)
```

```python
elem_one.drop(columns=['Unnamed: 0.1', 'Unnamed: 0'], inplace=True)
```

```python
elem_one.head()
```

```python
elem_one['X1'] = 1
elem_one['El2'] = np.nan
elem_one['X2'] = 0
elem_one['El3'] = np.nan
elem_one['El4'] = np.nan
elem_one['X3'] = 0
elem_one['X4'] = 0
elem_one.rename(columns={'_DateGap': '_date'}, inplace=True)
elem_one.head()
```

```python
unique_for_one = list(elem_one['El1'].unique())
unique_for_one
```

```python
dict_of_all_properties = {}

for i in unique_for_one:
    atomic_number = int(element_to_atomic_number.get(i))
    print(atomic_number)
    url = f"https://elementsapi.imet-db.ru/api/properties/allprops/{atomic_number}"
    response = requests.get(url)

    # Проверяем статус код ответа
    if response.status_code == 200:
        try:
            # Если запрос выполнен успешно, преобразуем ответ в JSON
            data = response.json()
            # Печатаем данные (или обрабатываем их по-другому)
            element_properties = []
            for item in data:
                if isinstance(item, dict):  # Убедитесь, что элемент является словарем
                    name = item.get("Name")
                    value = item.get("Value")
                    if name and value is not None:
                        element_properties.append((name, value))
            dict_of_all_properties[i] = element_properties
        except ValueError as e:
            # Если возникла ошибка при декодировании JSON, выводим сообщение об ошибке
            print(f"Error decoding JSON: {e}")
    else:
        # Если возникла ошибка, выводим статус код
        print(f"Error: {response.status_code}")
```

```python
# Создаем список новых имен колонок с суффиксом _1
new_column_names = [f"{prop_name}_1" for prop_name, _ in dict_of_all_properties[next(iter(dict_of_all_properties))]]

# Добавляем новые колонки в DataFrame и заполняем их значением None
for col_name in new_column_names:
    elem_one[col_name] = None
elem_one.head()
```

```python
for idx, row in elem_one.iterrows():
    element = row['El1']
    if element in dict_of_all_properties:
        for prop_name, prop_value in dict_of_all_properties[element]:
            elem_one.at[idx, prop_name] = prop_value
```

```python
elem_one_extended = elem_one
```
