### 1 - OHE representation of elements

```python
list(df)
```

```python
dff = df.drop(columns=['Unnamed: 0', 'SubstanceID','_DateSubstances', 'Direction', 'Comments',
                       'ReferenceID', '_DateGap', 'PrimaryKey', 'Modification', 'Elements', 'StructureType'])
dff.head()
```

```python
dff.isna().sum() / len(dff)
```

```python
dff['Temperature1'] = dff['Temperature1'].fillna(300.)
dff['Temperature2'] = dff['Temperature2'].fillna(300.)
```

```python
dff.isna().sum() / len(dff)
```

```python
dff['CrystalSystem'] = dff['CrystalSystem'].fillna('na')
```

```python
dff['CrystalSystem'] = dff['CrystalSystem'].apply(str.lower)
```

```python
dff['SpaceGroup'].value_counts()
```

```python
dff['SpaceGroup'] = dff['SpaceGroup'].fillna('na')
```

```python
dff[['El1', 'El2', 'El3', 'El4']] = dff[['El1', 'El2', 'El3', 'El4']].fillna('None')
dff[['X1', 'X2', 'X3', 'X4']] = dff[['X1', 'X2', 'X3', 'X4']].fillna(0.)
dff.isna().sum()
```

```python
dff['CrystalSystem']
```

```python
from sklearn.preprocessing import OneHotEncoder

dff['NumElements'] = dff['NumElements'].apply(str)
for i in range(1,5):
    dff[f'X{i}'] = dff[f'X{i}'].apply(int)
categorical = (dff.dtypes[dff.dtypes != np.float64]).index.tolist()
numeric = (dff.dtypes[dff.dtypes == np.float64]).index.tolist()

enc = OneHotEncoder(categories='auto', drop='first', sparse_output=False)
```

```python
enc
```

```python
#enc.fit_transform(dff.loc[:, categorical])

dff_categorical = enc.fit_transform(dff[categorical])
dff_categorical.shape
```

```python
for i in range(len(categorical)):
    print(categorical[i], len(enc.categories_[i]))
```

```python
print(dff_categorical.shape, dff[numeric].shape)

pd.DataFrame(dff_categorical, columns=enc.get_feature_names_out()).merge(dff[numeric].reset_index(drop=True),left_index=True, right_index=True)
```

```python
#dff_categorical.merge(dff[numeric])

dff_final = pd.DataFrame(dff_categorical, columns=enc.get_feature_names_out()).merge(dff[numeric].reset_index(drop=True),left_index=True, right_index=True)
dff_final.head()
```

```python
### baselines
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error as mse
from sklearn.metrics import root_mean_squared_error as rmse

X = dff_final.drop(columns=['E1', 'E2'])
y = dff_final[['E1']]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.3, random_state=42)

mean = np.mean(y_train)
ridge = Ridge(alpha=0.6).fit(X_train, y_train)
lasso = Lasso(alpha=0.01).fit(X_train, y_train)

pred_const = np.array([mean]*len(y_val))
pred_rid = ridge.predict(X_val)
pred_las = lasso.predict(X_val)

print('const rmse:', rmse(y_val, pred_const))
print('lasso rmse:', rmse(y_val, pred_las))
print('ridge rmse:', rmse(y_val, pred_rid))
```

```python
dff.iloc[y_val.index[np.where((y_val - pred_rid)**2 > 3)[0]],:].head(20)
```

```python
bst = GradientBoostingRegressor(n_estimators=200,
                                max_depth=5,
                                min_samples_split=10,
                                learning_rate=0.05,
                                verbose=1).fit(X_train, y_train)
pred_bst = bst.predict(X_val)

print(rmse(y_val, pred_bst))
```

```python
resid = (y_val - pred_rid)

plt.hist(resid, bins=30, color='blue', alpha=0.5)
plt.title('Model residuals')
plt.show()
```
